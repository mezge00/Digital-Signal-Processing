```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi as p
from thinkdsp import*
```

Generating The sampling frequency

```python
fm=eval(input("Enter the frequency "))
fs=eval(input("Enter the frequency "))
t=np.arange(0,0.05,0.0005)
t1=np.arange(0,0.05,(1/fs))
x=np.cos(2*p*fm*t)
xs=np.cos(2*p*t1*fm)
plt.plot(t,x,'g')
plt.stem(t1,xs,'r')
plt.show
plt.grid()

```

Amplitude Modulation

```python
Ac=eval(input("Enter the carrier signal amplitude "))
Am=eval(input("Enter the the message signal amplitude "))
fc=eval(input("enter the carrier signal frequency "))
fm=eval(input("enter the message signal frequency "))
t2=np.arange(-0.25,0.25,0.0001)
mu=Am/Ac
S=Ac*(1+mu*np.cos(2*p*fm*t2))*np.cos(2*p*fc*t2)
plt.plot(t2,S,'r')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('AM Under modulation')
plt.show
plt.grid()
```
